```python
def get_floor(input_str):
    input_list = list(input_str)
    floor = 0
    position = 0
    for i in input_list:
        position += 1
        if i == '(':
            floor += 1
        elif i == ')':
            floor += -1
        if floor < 0:
            return position
    return floor
```
